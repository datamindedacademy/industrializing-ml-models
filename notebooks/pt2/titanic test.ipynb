```python
from titanic import credentials, transformations
from titanic.storage import Storage
```

```python
import mlflow

mlflow.set_tracking_uri('http://127.0.0.1:5001')
```

## Load titanic test dataset

```python
credentials.load()
storage = Storage(bucket = "industrializing-ml-models")
```

```python
df_test = storage.read_csv('titanic/test.csv')
df_test = transformations.add_all_features(df_test)
```

### Instantiate classifier and make prediction for test data set

```python
model_name = "titanic-random-forest"
model_version = "latest"
model = mlflow.sklearn.load_model(f"models:/{model_name}/{model_version}")

test_features = transformations.select_features_for_ml(df_test)
predictions = model.predict(test_features)
df_test['Survived'] = predictions
df_test
```
